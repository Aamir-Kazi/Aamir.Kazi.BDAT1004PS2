###Question 3.

Write a class named Marsupial that can be used as shown below:

```python
class marsupial(object):
  def __init__(self):
    self.lst = []      
  def put_in_pouch(self,thing):
    self.lst.append(thing)
  def pouch_contents(self):
    return self.lst
        
m=marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents()
```

Now write a class named Kangaroo as a subclass of Marsupial that inherits all the
attributes of Marsupial and also:
- extends the Marsupial __init__ constructor to take, as input, the
coordinates x and y of the Kangaroo object,
- supports method jump that takes number values dx and dy as input and
moves the kangaroo by dx units along the x-axis and by dy units along the yaxis, and
- overloads the __str__ operator so it behaves as shown below.


```python
class kangaroo(marsupial):
  def __init__(self,x,y):
    marsupial.__init__(self)
    self.x=x
    self.y=y
  def jump(self,dx,dy):
    self.x+=dx
    self.y+=dy
  def __str__(self):
    return 'I am a kangaroo located at coordinates({},{})'.format(self.x,self.y)
    
k = kangaroo(0,0)
print(k)

k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())

k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

###Question 6.

Implement a class named HeadingParser that can be used to parse an HTML
document, and retrieve and print all the headings in the document. You should
implement your class as a subclass of HTMLParser, defined in Standard Library
module html.parser. When fed a string containing HTML code, your class should
print the headings, one per line and in the order in which they appear in the
document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your
implementation using w3c.html.


```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):

  def __init__(self):
    super().__init__()
    self.startTag = " "
        
  def handle_starttag(self, tag, attr):
    self.startTag = tag
        
  def handle_endtag(self, tag):
    self.startTag = ""
    
  def handle_data(self, data):
    if self.startTag == "h1":
      print(" " + data)
    elif self.startTag == "h2":
      print(" "*2 + data)
    elif self.startTag == "h3":
        print(" "*3 + data)
    elif self.startTag == "h4":
        print(" "*4 + data)
    elif self.startTag == "h5":
        print(" "*5 + data)
    elif self.startTag == "h6":
        print(" "*6 + data)
    else:
      pass

infile = open('w3c.html')
content = infile.read()
infile.close()
hp = HeadingParser()
hp.feed(content)
```

```python
from urllib.parse import urljoin 
from urllib.request import urlopen
from html.parser import HTMLParser

class Collector(HTMLParser):

  def __init__(self,url):
    super().__init__()
    self.url = url
    self.links = []
        
  def handle_starttag(self, tag, attrs):
    if tag == 'a':
      for attr in attrs:
        if attr[0] == 'href':
          absolute = urljoin(self.url, attr[1])
          if absolute[:4] == 'http':
            self.links.append(absolute)
                        
  def getLinks(self):
    return self.links
    

def webdir(url, depth, indent):
  if depth < 0:
    pass
    
  print("\t"*indent + url)
  resource = urlopen(url)
  content = resource.read().decode()
  collector = Collector(url)
  collector.feed(content)
  links = collector.getLinks()   
    
  for link in links:
    webdir(link, depth-1, indent+1)


webdir("http://reed.cs.depaul.edu/lperkovic/test1.html",2,0)
```
